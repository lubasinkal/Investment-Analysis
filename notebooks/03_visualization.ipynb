# Data Visualization for AAPL
## Introduction

In this notebook, we will create visualizations to better understand the historical stock data for AAPL. These visualizations will include the stockâ€™s price trends, moving averages, returns, volatility, and other relevant metrics.

### 1. Loading Data

We start by loading the processed data that was saved from the data analysis step.

```python
import pandas as pd

# Load the processed data
data = pd.read_csv('../data/processed/market data/processed_AAPL_data.csv', index_col=0)

# Display the first few rows of the dataset
data.head()
```

## 2. Price Trend Visualization
### 2.1 Closing Price Over Time

We visualize the closing price over the entire period to observe the general trend of the stock price

```python
import matplotlib.pyplot as plt

# Plot the closing price
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Closing Price', color='blue')
plt.title('Closing Price of AAPL Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Moving Averages
### 3.1 Short-term and Long-term Moving Averages

We calculate and plot short-term (30-day) and long-term (100-day) moving averages to smooth out short-term fluctuations and highlight longer-term trends.

```python
# Plot closing price with moving averages
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Closing Price', color='blue')
plt.plot(data['MA_30'], label='30-Day Moving Average', color='red')
plt.plot(data['MA_100'], label='100-Day Moving Average', color='green')
plt.title('AAPL Closing Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

Explanation

Moving averages help in identifying the direction of the trend. The 30-day moving average captures short-term trends, while the 100-day moving average highlights long-term trends.

## 4. Returns Visualization
### 4.1 Daily Log Returns

We visualize the daily log returns to understand the day-to-day percentage changes in the stock price.

```python
# Plot daily log returns
plt.figure(figsize=(12, 6))
plt.plot(data['Log Returns'], label='Log Returns', color='blue')
plt.title('Daily Log Returns of AAPL')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid(True)
plt.show()
```

### 4.2 Cumulative Returns

We visualize the cumulative returns to observe the total return over the entire period.

```python
# Plot cumulative returns
plt.figure(figsize=(12, 6))
plt.plot(data['Cumulative Percentage Return'], label='Cumulative Percentage Return', color='blue')
plt.title('Cumulative Percentage Return of AAPL')
plt.xlabel('Date')
plt.ylabel('Cumulative Return (%)')
plt.grid(True)
plt.show()
```

## 5. Volatility Visualization
### 5.1 Rolling Volatility

We plot the rolling volatility to understand the risk associated with the stock over different periods.

```python
# Plot rolling volatility
plt.figure(figsize=(12, 6))
plt.plot(data['Volatility 30'], label='30-Day Rolling Volatility', color='red')
plt.title('30-Day Rolling Volatility of AAPL')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.grid(True)
plt.show()
```

Explanation

Volatility measures the degree of variation in the stock's returns. Higher volatility indicates greater risk, and this plot shows how volatility changes over time

## 6. Combined Visualization
### 6.1 Log Returns and Volatility

We plot log returns and volatility together to understand their relationship.

```python
# Plot log returns and volatility
plt.figure(figsize=(14, 7))

# Plot log returns
plt.subplot(2, 1, 1)
plt.plot(data['Log Returns'], label='Log Returns', color='blue')
plt.title('Log Returns for AAPL')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid(True)
plt.legend()

# Plot volatility
plt.subplot(2, 1, 2)
plt.plot(data['Volatility 30'], label='Volatility (30-Day Rolling Std)', color='red')
plt.title('Volatility for AAPL')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## Explanation

Plotting log returns and volatility together helps in understanding how changes in returns correspond with changes in risk. This combined view can highlight periods of high volatility and their impact on returns.

## Conclusion

In this notebook, we visualized various aspects of the historical stock data for AAPL. These visualizations help in understanding the stock's price trends, moving averages, daily returns, cumulative returns, and volatility. Such insights are crucial for making informed investment decisions.
